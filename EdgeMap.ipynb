## 엣지맵 추출

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

# CLAHE 생성
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# 결과 저장 디렉토리 설정
output_dir = './dataset/ghoumrassi/ghoumrassi_edge_new'
os.makedirs(output_dir, exist_ok=True)


for index, img in enumerate(images):
    # 이미지 파일 읽기 (그레이스케일)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_img = cv2.GaussianBlur(gray_img, (7, 7), 0)

    # CLAHE 적용
    clahe_img = clahe.apply(gray_img)

    # 히스토그램 평활화
    # equalized_image = cv2.equalizeHist(clahe_img)

    # 에지 검출
    edges = cv2.Canny(clahe_img, 150, 200)
    edges_inverted = cv2.bitwise_not(edges)

    # 결과 저장
    # new_index = index + 991  # 인덱스에 991 더하기
    edge_filename = os.path.join(output_dir, f'edge_{index}.jpg')
    cv2.imwrite(edge_filename, edges_inverted)

    # edge_filename = os.path.join(output_dir, f'edge_{index}.jpg')
    # cv2.imwrite(edge_filename, edges_inverted)

    # # 결과 이미지 확인 (선택 사항)
    # plt.figure(figsize=(10, 5))
    # plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    # plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    # # plt.subplot(132), plt.imshow(edges_inverted, cmap='gray')
    # # plt.title('Equalized Image'), plt.xticks([]), plt.yticks([])
    # plt.subplot(133), plt.imshow(edges_inverted, cmap='gray')
    # plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    # plt.show()
```

```python
import os
import json
import re

# JSON 파일들이 저장된 디렉토리 경로 설정
json_dir = './dataset/karol-skorulski/karol-skorulski_text'

# "white"가 포함된 파일명의 숫자를 저장할 리스트
white_numbers = []
keywords = ["white", "bright", "light", "pale", "ivory", "cream", "sky", "gray"]

# 디렉토리 내의 모든 JSON 파일에 대해 반복
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):  # JSON 파일만 처리
        file_path = os.path.join(json_dir, filename)
        with open(file_path, 'r') as json_file:
            data = json.load(json_file)
            text = data.get('text', "").lower()  # 대소문자 구분 없이 텍스트 가져오기
            # 텍스트에서 특정 단어들이 포함된 경우 파일명에서 숫자 추출
            if any(keyword in text for keyword in keywords):
                numbers = re.findall(r'\d+', filename)  # 파일명에서 숫자 추출
                if numbers:  # 숫자가 추출된 경우에만
                    white_numbers.append(int(numbers[0]))  # 첫 번째 숫자만 사용

# 결과 출력
white_numbers.sort()

print("Numbers extracted from files containing 'white':")
for number in white_numbers:
    print(number)
```

## 최종 엣지맵 추출

```python
####### 화이트 이미지에 대해서는 (히스토그램 평활화) 처리 #######
import cv2
import os

def white_edge(img, clahe, output_dir, number):
    # 그레이스케일 변환
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_img = cv2.GaussianBlur(gray_img, (3, 3), 0)

    # CLAHE 적용
    clahe_img = clahe.apply(gray_img)

    # 히스토그램 평활화
    equalized_image = cv2.equalizeHist(clahe_img)

    # 에지 검출
    edges = cv2.Canny(equalized_image, 100, 300)
    edges_inverted = cv2.bitwise_not(edges)

    # 결과 저장
    edge_filename = os.path.join(output_dir, f'edge_{number}.jpg')
    # cv2.imwrite(edge_filename, edges_inverted)

    # print(f"(white)Processed and saved: {edge_filename}")
    
    plt.subplot(133), plt.imshow(edges_inverted, cmap='gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
####### 기본 코드 (canny edge) #######
def canny_edge(img, clahe, output_dir, number):
    # 그레이스케일 변환
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # CLAHE 적용
    clahe_img = clahe.apply(gray_img)

    # 에지 검출
    edges = cv2.Canny(clahe_img, 100, 300)
    edges_inverted = cv2.bitwise_not(edges)

    # 결과 저장
    edge_filename = os.path.join(output_dir, f'edge_{number}.jpg')
    cv2.imwrite(edge_filename, edges_inverted)

    print(f"Processed and saved: {edge_filename}")
```

```python
def white_edge(img, clahe, output_dir, number):
    # 그레이스케일 변환
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_img = cv2.GaussianBlur(gray_img, (5,5), 0)

    # CLAHE 적용
    clahe_img = clahe.apply(gray_img)

    # 히스토그램 평활화
    # equalized_image = cv2.equalizeHist(clahe_img)

    # 에지 검출
    edges = cv2.Canny(clahe_img, 50, 100)
    edges_inverted = cv2.bitwise_not(edges)

    # 결과 저장
    edge_filename = os.path.join(output_dir, f'edge_{number}.jpg')
    cv2.imwrite(edge_filename, edges_inverted)

    print(f"(white)Processed and saved: {edge_filename}")
    
    # plt.subplot(133), plt.imshow(edges_inverted, cmap='gray')
    # plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    # plt.show()


# 사용 예시
# 이미지 디렉토리 및 CLAHE 객체 생성
image_dir = './dataset/karol-skorulski/karol-skorulski_image'
output_dir = './dataset/karol-skorulski/karol-skorulski_edge_new'
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# 디렉토리 내의 모든 이미지 파일 처리
for filename in os.listdir(image_dir):
    if filename.startswith('img_') and filename.endswith('.jpg'):
        number = filename.split('_')[1].split('.')[0]
        
        if int(number) in white_numbers:
            img_path = os.path.join(image_dir, filename)
            img = cv2.imread(img_path)

            if img is not None:
                white_edge(img, clahe, output_dir, number)
                # break
        else:
            continue
            
```
